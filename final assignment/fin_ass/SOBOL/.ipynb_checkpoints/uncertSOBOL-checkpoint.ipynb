## 7)Visualize the relation between objectives and Uncertainty

```python
#List of uncertain Parameters
uncertain_para = ['A.0_ID flood wave shape', 'A.1_Bmax', 'A.1_Brate', 'A.1_pfail',
       'A.2_Bmax', 'A.2_Brate', 'A.2_pfail', 'A.3_Bmax', 'A.3_Brate',
       'A.3_pfail', 'A.4_Bmax', 'A.4_Brate', 'A.4_pfail', 'A.5_Bmax',
       'A.5_Brate', 'A.5_pfail', 'discount rate 0', 'discount rate 1',
       'discount rate 2']

#List of objectives
objectives = ['A.1_Expected Annual Damage',
       'A.2_Expected Annual Damage', 'A.3_Expected Annual Damage',
       'A.4_Expected Annual Damage', 'A.5_Expected Annual Damage',
       'A.1_Expected number of deaths', 'A.2_Expected number of deaths',
       'A.3_Expected number of deaths', 'A.4_Expected number of deaths',
       'A.5_Expected number of deaths']
#Create new dataframe containing  correlation coefficients between uncertainties and objectives
corr_df = pd.DataFrame(columns = uncertain_para,index=objectives)
for i in uncertain_para:
    for j in objectives:
        corr_df.loc[j,i] = final_df[j].astype(float).corr(final_df[i].astype(float))
#Visualize the results in a heatmap
plt.figure(figsize=(16, 8))
sns.heatmap(corr_df.astype(float),annot=True,cmap="viridis")
```

## 8)Perform linear regression for 2 main objectives of Total annual expected damage and Total number of expected deaths at all 5 locations

```python
df_agg = pd.DataFrame()
#Aggregate the reults over 5 locations
df_agg['Total annual expected damage'] = results_agg['A.1_Expected Annual Damage']+ results_agg['A.2_Expected Annual Damage']+ results_agg['A.3_Expected Annual Damage']+ results_agg['A.4_Expected Annual Damage']+ results_agg['A.5_Expected Annual Damage'] 
                                         
df_agg['Total number of expected deaths'] = results_agg['A.1_Expected number of deaths'] + \
                                            results_agg['A.2_Expected number of deaths'] + \
                                            results_agg['A.3_Expected number of deaths'] + \
                                            results_agg['A.4_Expected number of deaths'] + \
                                            results_agg['A.5_Expected number of deaths']
```

### 8.1.Linear regression for expected damage

```python
y1 = df_agg['Total annual expected damage'].values
model1 = sm.OLS(y1,uncertain.astype(float))
results = model1.fit()
print(results.summary())                                    
```

### 8.2.Linear regression for expected number of deaths

```python
y2 = df_agg['Total number of expected deaths'].values
model2 = sm.OLS(y2,uncertain.astype(float))
results = model2.fit()
print(results.summary())
```

### 8.3. Perform linear regressions with only Probability of failure:Total Expected Damage

```python
failure_prob = uncertain[['A.1_pfail','A.2_pfail','A.3_pfail','A.4_pfail','A.5_pfail']]
y3 = df_agg['Total annual expected damage'].values
model3 = sm.OLS(y3,failure_prob.astype(float))
results = model3.fit()
print(results.summary())        
```

### 8.4. Perform linear regressions with only Probability of failure:Expected number of deaths

```python
y4 = df_agg['Total number of expected deaths'].values
model4 = sm.OLS(y4,failure_prob.astype(float))
results = model4.fit()
print(results.summary())   
```
